# Defining the model

```python
import tensorflow.keras as keras
from keras.utils.vis_utils import plot_model
from keras.layers import  BatchNormalization
```

```python
comparator = keras.models.Sequential()
comparator.add(keras.layers.Conv2D(kernel_size=(3,3), filters = 128, input_shape=(None, None, 3), activation='relu', padding = "same"))
comparator.add(keras.layers.Conv2D(kernel_size=(3,3), filters = 64, activation='relu', padding = "same"))
comparator.add(keras.layers.Conv2DTranspose(filters = 3, kernel_size = (3,3), output_padding = 1, padding = "same", stride = 2, activation = "relu"))
# comparator.add(keras.layers.Conv2D(kernel_size=(3,3), filters = 3, activation='relu', padding = "same"))
```

```python
plot_model(comparator, show_shapes=True)
```

```python
comparator.compile(loss=["mse"], optimizer=keras.optimizers.Adam(learning_rate=0.001,
    beta_1=0.9,
    beta_2=0.999,
    epsilon=1e-07,
    amsgrad=False,
    name="Adam"))
```

# Training the model

```python
from keras.callbacks import LearningRateScheduler
```

```python
def step_decay(epoch, lr):
  initial_lr = 1e-3
  if epoch<30:
    lr = initial_lr
  else:
    lr = lr/10
    
  return lr
```

```python
checkpoint_path = "/content/gdrive/MyDrive/Models/SRCNN/"
checkpoint_dir = os.path.dirname(checkpoint_path)

# Create a callback that saves the model's weights
cp_callback = keras.callbacks.ModelCheckpoint(filepath=checkpoint_path,
                                                 save_weights_only=True,
                                                 verbose=1)
```

```python
lrs = LearningRateScheduler(step_decay)
```

```python
comparator.load_weights(checkpoint_path)
```

```python
print(noisy_patches.shape, patches.shape)
```

```python
comparator.fit(noisy_patches, patches, batch_size = 128, epochs = 50, callbacks = [lrs, cp_callback], verbose = 1)
```
